```python
import matplotlib.pyplot as plt

jenis = ('Iphone 16 Pro','Galaxy S24','Xiaomi 14 Ultra','Huawei P50')
penjualan = (4000,5000,1500,1300)
warna = ('#8B4513','#FFF8DC','#93C572','#E67F0D')
explode = (0.1,0,0,0)
plt.title('Penjualan Handphone')
plt.pie(penjualan, labels=jenis, autopct='%1.2f%%', colors=warna, explode=explode)
plt.axis('equal')  # To ensure the pie chart is a circle
plt.show()
```
